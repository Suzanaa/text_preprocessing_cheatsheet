```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,11)
y = x**2
```

```python
x
```

## Matplotlib

```python
plt.plot(x,y, 'o')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(x,y,"b")

plt.show()
```

## Seaborn

```python
import seaborn as sns
%matplotlib inline
```

## Heatmaps

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
f = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
f # interpret: e.g. in January of 1950 there were 115 passengers
```

```python
sns.heatmap(f,cmap='magma')
```

```python
sns.heatmap(f,cmap='coolwarm', linecolor='white', linewidth=1)
```

## Cluster map (hierarchical clustering)

```python
sns.clustermap(f)
```

```python
sns.clustermap(f,cmap='coolwarm', standard_scale=1)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique() # checking unique labels
```

```python
sns.pairplot(iris)
```

## Style and color

```python
sns.set_style('ticks')
sns.countplot(x='species',data=iris)
sns.despine()
```

```python
sns.lmplot(x='year',y='passengers',data=flights,hue='month') # change colors with palette='magma' etc.
```

## Pandas

```python
data.dropna()
data.head()
```

```python
data.plot.line(x = data.date ,y = 'open', figsize=(12,3))
data.plot.line(x = data.date ,y = 'close', color='r', figsize=(12,3))
data.plot.line(x = data.date ,y = 'volume', color='g', figsize=(12,3))
```
