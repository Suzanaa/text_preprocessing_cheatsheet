```python
columns = [lexicon['AFRAID'], lexicon['AMUSED'], lexicon['ANGRY'], lexicon['ANNOYED'], lexicon['DONT_CARE'], lexicon['HAPPY'], lexicon['INSPIRED'], lexicon['SAD'], lexicon['freq']]

for item in columns:
    print(item.name)
    print('min: %s' %np.min(item))
    print('max: %s' %np.max(item))
    print('mean: %s' %np.mean(item))
    print('mean: %s' %np.median(item))
    print('____________________________')
    print(' ')
```
